#ANA fazendo LIMPEZA

```python
gastos_limpos = gastos
faturamento_limpo = faturamento
print(gastos_limpos.Tipo_gasto.unique())
```

```python
gastos_limpos['Tipo_gasto'] = gastos_limpos['Tipo_gasto'].replace(['Campanha Facebook', 'Campanha facebook', 'Campanha Facebok', 'Campanha Facebook '], 'Facebook')
gastos_limpos['Tipo_gasto'] = gastos_limpos['Tipo_gasto'].replace(['Contratação GoogleAds','Contratação GoogleAdds','Contratação  GoogleAds','Contratação GoogleAds '],'Google_Adds')
gastos_limpos['Tipo_gasto'] = gastos_limpos['Tipo_gasto'].replace(['Campanha Instagran', 'Campanha Instagram', 'Campanha instagram'],'Instagram')
gastos_limpos['Tipo_gasto'] = gastos_limpos['Tipo_gasto'].replace(['Contratação Propagandas Outdoor', 'Contratação Propagandas Outdor ' ,'Contratação Propagandas Outdoor '],'Outdoor')
gastos_limpos['Tipo_gasto'] = gastos_limpos['Tipo_gasto'].replace(['Contratação Propaganda Rádio ', 'Contratação Propaganda Radio','Contratação Propaganda Rádio','Contratação Propaganda Radio '], 'Radio')
gastos_limpos['Tipo_gasto'] = gastos_limpos['Tipo_gasto'].replace(['Contratacão Propaganda Revistas', 'Contratação Propaganda Revistas', 'Contratação Propaganda Revista','Contratação Propaganda Revistas '],'Resvistas')
gastos_limpos['Tipo_gasto'] = gastos_limpos['Tipo_gasto'].replace(['Campanha Tiktok', 'Campanha Tik-tok', 'Campanha Tik tok', 'Campanha tiktok'],'TikTok')
gastos_limpos['Tipo_gasto'] = gastos_limpos['Tipo_gasto'].replace(['Contratação Propaganda tv', 'Contratação Propaganda TV','Contratação Propaganda televisão'],'Televisao')
gastos_limpos['Tipo_gasto'] = gastos_limpos['Tipo_gasto'].replace(['Campanha Youtube'], 'YouTube')

gastos_limpos.to_excel('dados_limpos_tipo.xlsx', index=False)
ursin.read_excel('/content/dados_limpos_tipo.xlsx')#só verificando se salvou certinho
print(gastos_limpos.Tipo_gasto.unique())
```

#ANA fazendo PRÉ-PROCESSAMENTO

```python
mes = gastos_limpos['Data Faturamento'].dt.month.astype(str).str.zfill(2)
anos = gastos_limpos['Data Faturamento'].dt.year.astype(str)
gastos_limpos["Ano_Mes"] = anos + "-" + mes
gastos_limpos["Ano_Mes"] = ursin.to_datetime(gastos_limpos["Ano_Mes"])

gastos_empresas = gastos_limpos.groupby("Empresa").groups
gastos_mensais = gastos_limpos.groupby("Ano_Mes").sum(numeric_only=True).sort_values("Ano_Mes")


print(gastos_limpos.head())

indx = gastos_mensais.index.tolist()

print(indx)

gastos_mensais.to_excel('gastos_por_mes.xlsx', index=False)
```

```python
gastos_limpos.set_index('Ano_Mes', inplace=True)
```

```python
print(gastos_mensais.info())
indices = gastos_mensais.index.tolist()

print(indices)
```

```python
mes = faturamento_limpo['Mes_Ano'].dt.month.astype(str).str.zfill(2)
anos = faturamento_limpo['Mes_Ano'].dt.year.astype(str)
faturamento_limpo["Ano_Mes"] = anos + "-" + mes
faturamento_limpo['Ano_Mes'] = ursin.to_datetime(faturamento_limpo['Ano_Mes'])

#faturamento_limpo

faturamento_mensal = faturamento_limpo.groupby("Ano_Mes").sum(numeric_only=True).sort_values('Ano_Mes')
faturamento_mensal.to_excel('faturmento_por_mes.xlsx', index=True)
```

```python
faturamento_mensal['Custo'] = gastos_mensais['Valor']
print(faturamento_mensal.head())

indices = faturamento_mensal.index.tolist()

print(indices)
```

```python
faturamento_mensal['Relacao_F/C'] = faturamento_mensal['Faturamento'] / faturamento_mensal['Custo']
faturamento_mensal.head()
```


#FIM da ANA fazendo PRÉ-PROCESSAMENTO

```python
faturamento_mensal['Relacao_F/C'].plot(title = 'Relação F/C mensal', kind='line', color ='purple',)#uma visualização do faturamento

graficar.xlabel("Ano_Mês")
graficar.ylabel("Índice de Retorno")
```

```python
faturamento_mensal['F/C_normalizada'] = (faturamento_mensal['Relacao_F/C'] - faturamento_mensal['Relacao_F/C'].min()) / (faturamento_mensal['Relacao_F/C'].max() - faturamento_mensal['Relacao_F/C'].min())
faturamento_mensal.head()
```

```python
fig = graficar.figure(figsize = (8,5))
f = faturamento_mensal['Relacao_F/C']
eixo = fig.add_axes([0,0,1,1])
eixo.plot(faturamento_mensal.index,f, color = 'blue')
eixo.set_title('Relação F/C', fontsize = 25, pad = 20)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Taxa de faturamento', fontsize=20)
eixo.axhline(max(f), color = 'k', linestyle='--')
eixo.axhline(f.mean(), color = 'k', linestyle='--')
# Coloca uma flecha que aponta para a linhda do valor médio
eixo.annotate('Valor médio', xy=(3, f.mean()), xytext=(5, f.mean()+1.5),
              arrowprops=dict(facecolor='black', shrink=0.05))
eixo.axhline(min(f), color = 'k', linestyle='--')
eixo.grid(True)
```

```python
indice_min = faturamento_mensal.index[faturamento_mensal['Relacao_F/C'] == min(f)].tolist()

##correspondente = faturamento_mensal.loc[faturamento_mensal['WXY'] == valor, 'ABC'].values[0]

print(indice_min)
```

```python
empresa_min = gastos_limpos.loc[indice_min, ['Empresa', 'Tipo_gasto','Valor']]
# Agrupar por 'Empresa' e os 'Valores'
emp_min_emp = empresa_min.groupby('Empresa').sum(numeric_only=True).sort_values('Valor')
#Agrupa por 'Tipo de gasto' e Valor
emp_min_tp = empresa_min.groupby('Tipo_gasto').sum(numeric_only=True).sort_values('Valor')
emp_min_tp.head()
```

```python
indice_max = faturamento_mensal.index[faturamento_mensal['Relacao_F/C'] == max(f)].tolist()
empresa_max = gastos_limpos.loc[indice_max, ['Empresa', 'Tipo_gasto', 'Valor']]
# Agrupar por 'Empresa' e os 'Valores'
emp_max_emp = empresa_max.groupby('Empresa').sum(numeric_only=True).sort_values('Valor')
#Agrupo por 'Tipo de gasto' e 'Valor'
emp_max_tp = empresa_max.groupby('Tipo_gasto').sum(numeric_only=True).sort_values('Valor')
emp_max_tp
```
